```python
import numpy as np
from matplotlib import pyplot as plt
import mpl_toolkits.mplot3d

w0 = 1.0
f = 1.0
w, b = np.mgrid[0.1: 3: 359j, 0.1: 3: 359j]
Ax = f*((w0 ** 2-w**2)**2+4*b**2*w**2)**(-0.5)
font2 = {'family' : 'Times New Roman','weight' : 'normal','size' : 15,}    
pw_vec = np.linspace(0.1, 0.98, 1000)
pb_vec = ((1-pw_vec**2)/2)**0.5
pA_vec = (1-pw_vec**4)**(-0.5)
plt.figure(figsize=(12, 8))
ax = plt.subplot(111, projection = '3d')
ax.view_init(elev=30, azim = 60)
ax.plot_surface(w, b, Ax, cmap=plt.cm.coolwarm)
ax.plot(pw_vec, pb_vec, pA_vec, 'd-',linewidth = 5.0, label='peak')
ax.set_zlim(0, 2.8)
ax.set_xlabel(r'$\frac{\omega}{\omega_0}$', font2)
ax.set_ylabel(r'$\frac{\beta}{\omega_0}$', font2)
ax.set_zlabel(r'$\frac{A\omega_0^2}{f}$', font2)
ax.set_title(r'A vs $\omega$ and $\beta$')
ax.legend()
plt.show()
```

```python
import matplotlib.style as psl
psl.use('seaborn-paper')

w_vec = np.linspace(0.01, 3.0, 100)
b_vec = np.array([0.25*2**i for i in range(6)])

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].plot(w_vec, f*((w0 ** 2-w_vec**2)**2)**(-0.5), label = r'$\beta=0.0$')
ax[1].plot(w_vec, f*((w0 ** 2-w_vec**2)**2)**(-0.5)*w_vec, label = r'$\beta =0$')
for i in range(6):
    A_vec = f*((w0 ** 2-w_vec**2)**2+4*b_vec[i]**2*w_vec**2)**(-0.5)
    v_vec = A_vec*w_vec
    ax[0].plot(w_vec, A_vec, label = r'$\beta='+str(b_vec[i])+'$')
    ax[1].plot(w_vec, v_vec, label = r'$\beta='+str(b_vec[i])+'$')

font2 = {'family' : 'Times New Roman','weight' : 'normal','size' : 15,}
w1_vec = np.linspace(0.2, 0.97, 10)
ax[0].plot(w1_vec, (1-w1_vec**4)**(-0.5), 'b:')
ax[1].vlines(1.0, 0.0, 2.5, linestyles='dotted', colors = 'b')
ax[0].set_title(r'A vs $\omega$')
ax[1].set_title(r'$A_v$ vs $\omega$')
ax[0].set_xlabel(r'$\frac{\omega}{\omega_0}$', font2)
ax[1].set_xlabel(r'$\frac{\omega}{\omega_0}$', font2)
ax[0].set_ylabel(r'$\frac{A\omega_0^2}{f}$', font2)
ax[1].set_ylabel(r'$\frac{A_v\omega_0^3}{f}$', font2)
ax[0].set_ylim(0, 2.5)
ax[1].set_ylim(0, 2.5)
ax[0].legend()
ax[1].legend()
ax[0].grid('on')
ax[1].grid('on')
```

```python
psl.use('seaborn-paper')
font1 = {'family' : 'Times New Roman','weight' : 'normal','size' : 13,}
font2 = {'family' : 'Times New Roman','weight' : 'normal','size' : 15,}
w0 = 1.0
w_vec = np.linspace(0.01, 2.0, 138)
b_vec = np.array([0.25*2**i for i in range(6)])

plt.figure(figsize=(8,7))
ax = plt.gca()
for i in range(6):
    phi_vec = np.arctan(-2*b_vec[i]*w_vec/(w0**2-w_vec**2))+np.sign(1-w_vec)*np.pi/2
    plt.plot(w_vec, phi_vec, label = r'$\beta='+str(b_vec[i])+'$')

ax.set_title(r'Phase angle $\phi$ vs $\omega$', font2)
ax.grid('on')
ax.set_xlabel(r'$\frac{\omega}{\omega_0}$', font2)
ax.set_ylabel(r'$\phi_v$', font1)
yto2 = np.linspace(-np.pi/2, np.pi/2, 7)
yto1 = np.linspace(-np.pi, 0, 7)
xto2 = np.linspace(0.1,1.9, 7)
yt2 = [r'$-\frac{\pi}{2}$',r'$-\frac{\pi}{3}$',r'$-\frac{\pi}{6}$',r'$0$',r'$\frac{\pi}{6}$',r'$\frac{\pi}{3}$',r'$\frac{\pi}{2}$']
yt1 = [r'$-\pi$',r'$-\frac{5\pi}{6}$',r'$-\frac{2\pi}{3}$',r'$-\frac{\pi}{2}$',r'$-\frac{\pi}{3}$',r'$-\frac{\pi}{6}$',r'$0$']
xt2 = [r'$0.1$', r'$0.4$', r'$0.7$', r'$1.0$', r'$1.3$',r'$1.6$',r'$1.9$']
plt.yticks(yto2, yt2)
plt.xticks(xto2, xt2)
plt.tick_params(labelsize=15)
ax2 = ax.twinx()
ax2.set_ylabel(r'$\phi$', font1)
ax2.vlines(0, -np.pi, 0,alpha=0.0)
plt.yticks(yto1, yt1)
plt.tick_params(labelsize=15)
ax.legend()
plt.show()
```
