```python
import numpy as np
from matplotlib import pyplot as plt

font1 = {'family' : 'Times New Roman','weight' : 'normal','size' : 14,}   
font2 = {'family' : 'Times New Roman','weight' : 'normal','size' : 20,}   
fig, ax = plt.subplots(1, 2, figsize=(10, 4.5))
for i in range(2):
    ax[i].spines['right'].set_color('none')
    ax[i].spines['top'].set_color('none')
    ax[i].spines['left'].set_position(('data', 0))
    ax[i].spines['bottom'].set_position(('data', 0))
    ax[i].set_yticks([])
    ax[i].set_xticks([])

t = np.linspace(0, 20, 300)
xt = np.sin(t)
vt = np.cos(t)
ax[0].plot(t, xt)
ax[1].plot(xt, vt)
ax[0].set_title(r'$x$ v.s. $t$', font2)
ax[1].set_title(r'$x$ v.s. $p$', font2)
ax[0].set_xlabel(r'$t$', font1)
ax[0].set_ylabel(r'$x$', font1)
ax[0].set_ylim([-1.5, 1.5])
ax[1].set_xlim([-1.3, 1.3])
ax[1].set_ylim([-1.3, 1.3])
ax[1].set_xlabel('x', font1, horizontalalignment='right', position=(0.92, -0.2))
ax[1].set_ylabel('p', font1, horizontalalignment='right', position=(-0.2, 0.92))

ax[1].arrow(xt[30], vt[30], xt[32]-xt[30], vt[32]-vt[30], fc='r', overhang=0.5, head_width = 0.1)

plt.show()
```
